```python
# Job Title을 XXX Anaiyst로 통일시켜 줄 것임
# 예시) Marketing Analyst/ Revenue Analyst/ Data Analyst 등 
# 우선 뒤에 at 붙은 것 삭제해주어야 겠다. 
```

```python
Title = []
total = len(job_list)

for i in range(total):
    k = job_list[i].partition(' at ')[0]
    Title.append(k)
    
Title
```

```python
len(set(Title))
```

```python
# JobTitle 이 822개여서 더 줄여야 함.
# 우선 JobTitle에서 'analyst'가 포함되지 않는 것을 살펴볼 예정

non_analyst = []
for i, job in enumerate(Title):
    if 'analyst' not in job:
        non_analyst.append(i)
```

```python
len(non_analyst)
```

```python
key_vocab = ['marketing', 'revenue', 'business', 'financial', 'data', 'investment banking', 'pricing', 'securities', 'system', 'consulting', '']
```

```python
# 우선 위에 것 지우고 시작해야겠다. 
check = []
del_list = []

for i, job in enumerate(Title):
    for key in key_vocab:
        check.append(key in job)
    if 'analyst' in job and any(check):
        pass
    else:
        del_list.append(i)
    del check[:]
print(del_list)
len(del_list)
```
